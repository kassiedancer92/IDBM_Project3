## EDA of movie data

```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
#Load in data from 2000
df_2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_2000.head()
```

```python
#Load in data from 2001
df_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
df_2001.head()
```

```python
#Concatenate into one dataframe
df = pd.concat((df_2000, df_2001))
df.info()
```

## Questions:
1. How many movies had at least some valid financial information (values > 0 for budget OR revenue)?
    - Please exclude any movies with 0's for budget AND revenue from the remaining visualizations.
2. How many movies are there in each of the certification categories (G/PG/PG-13/R)?
3. What is the average revenue per certification category?
4. What is the average budget per certification category?

##### Question 1
    There are 643 movies with some financial information 

```python
#Filter out movies with a budget or revenue listed
some_financial = df[(df['budget'] > 0) | (df['revenue'] > 0)]
some_financial
```

```python
#Filter out movies with 0 for budget and revenue
df_final = df[(df['budget'] > 0) & (df['revenue'] > 0)]
df_final
```

#### Question 2
How many movies are there in each of the certification categories (G/PG/PG-13/R)?
- G: 13
- PG: 30
- PG-13: 117
- R: 143


```python
#Find the counts of movies in each category
value_counts = df_final['certification'].value_counts()
value_counts
```

```python
#Create a countplot of certification categories
ax = sns.countplot(data= df_final, x= 'certification', order = value_counts.index)
ax.set_title('Movies by Certifications');
```

### Question 3
What is the average revenue per certification category?

-  G  $133,216,900

- NR  $43,133,930

- PG  $ 129,125,600

- PG-13 $ 111,101,800

- R $ 47,884,240

```python
#Group revenue and certification to find average revenue
avg_revenue = df_final.groupby('certification')['revenue'].mean()
avg_revenue
```

```python
#Create barplot of average revenue by certification
ax = sns.barplot(data = df_final, x= 'certification', y= 'revenue', order = value_counts.index)
ax.set_title('Average Revenue by Certification');
```
